##### a) Utilizando a matriz Numpy, construa um dicionário que conta quantos gols cada país já fez em copas

````python
# SOLUÇÃO

m = df[['HomeTeamName','HomeTeamGoals','AwayTeamName','AwayTeamGoals']].values
gols={}

for p1, g1, p2, g2 in m:
    
    if p1 not in gols:
        gols[p1] = g1
    else:
        gols[p1] +=g1

    if p2 not in gols:
        gols[p2] = g2
    else:
        gols[p2] +=g2
gols
````

##### b) Utilizando a matriz Numpy, construa um dicionário que conta o público (Attendance) total que o país já teve em copas.

````python
# SOLUÇÃO
listapublico = df[['HomeTeamName','AwayTeamName','Attendance']].values
attend={}
for p1, p2, publico in listapublico:
    
    if p1 not in attend:
        attend[p1] = publico
    else:
        attend[p1] += publico
        
    if p2 not in attend:
        attend[p2] = publico
    else:
        attend[p2] += publico 
        
attend 
````

#### c) Insira os dois dicionários acima em outro dicionário para criar um DataFrame onde cada dicionário representa uma coluna. Salve o DataFrame em um arquivo .csv.

````python
# SOLUÇÃO
df = pd.DataFrame({'Goals': gols, 'Attendance' : attend})
df.to_csv('samanta.csv', index=False)
df
````
